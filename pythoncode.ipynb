```python
import pandas as pd

df = pd.read_csv(r'dataset.csv')
print(df)
```

```python
index=0
training_x=[]
testing_x=[]
for i in df['area']:
    if index<500:
        training_x.append(i)
    else:
        testing_x.append(i)

    index=index+1

    
```

```python
j=0
training_y=[]
testing_y=[]
for i in df['price']:
    if j<500:
        training_y.append(i)
    else:
        testing_y.append(i)
    j=j+1
```

```python
a=0
b=0
index=0
alpha = 0.00000001
m=len(training_x)
while(index<1000):
    sum=0
    sum_a=0
    sum_b=0
    for i in range (0,m):
        h = a + b*training_x[i]
        sum=sum+((h-training_y[i])*(h-training_y[i]))
        sum_a += a + b * training_x[i] - training_y[i]
        sum_b += (a + b * training_x[i] - training_y[i]) * training_x[i]
    J=(1/2*m)*sum
    derivate_a = (1/m)*(sum_a)
    derivate_b = (1/m)*(sum_b)

    a=a- (alpha*derivate_a)
    b=b- (alpha*derivate_b)
    index=index+1
```

```python
test=int(input('Enter the area'))
```

```python
print(a+b*test)
```

```python
import numpy as np
predicted_y=[]
for value in range (0,500):
    predicted_y.append(a+(b*training_x[value]))
    
```
